```python
# import required Library
import apriori
import pandas as pd
import numpy as np
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
```

```python
# load the dataset
df=pd.read_csv("/content/Market_Basket_Optimisation.csv",encoding="cp1252")
df.head(5)
```

```python
# deal with missing values
df.fillna(0,inplace=True)
```

```python
df.head(10)
```

```python
#for using aprori need to convert data in list format..
# transaction = [['apple','almonds'],['apple'],['banana','apple']]....
# Convert the Pandas DataFrame to a list of lists format.
transactions = []
for i in range(0,len(df)):
    transactions.append([str(df.values[i,j]) for j in range(0,20) if str(df.values[i,j])!='0'])
    transactions.append(transactions)
```

```python
transactions[3]
```

```python
# Create a TransactionEncoder object.
encoder = TransactionEncoder()

# Fit the TransactionEncoder object to the list of lists.
# encoder.fit(transactions)

# Transform the list of lists into a one-hot encoded array.
onehot = encoder.transform(transactions)

# Convert the one-hot encoded array to a Pandas DataFrame.
onehot = pd.DataFrame(onehot, columns=encoder.columns)

# Create an Apriori object.
apriori = apriori(min_support=0.01, min_confidence=0.5)

# Fit the Apriori object to the Pandas DataFrame.
apriori.fit(onehot)

# Generate association rules from the Apriori object.
association_rules = apriori.rules

# Print the association rules.
for rule in association_rules:
  print(rule)
```
